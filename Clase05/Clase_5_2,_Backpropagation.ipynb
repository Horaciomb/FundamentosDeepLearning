### Parte A, Realizamos la gradiente de manera manual.

Pista: no necesitas ninguna libreria

```python
a = 1
b = 2
c = 3
```

```python
d = a + b
e = b - c
```

```python
y = d*e
```

```python
y
```

```python
dy_dd = e
dy_de = d
```

```python
dd_da = 1
dd_db = 1
```

```python
de_db = 1
de_dc = -1
```

```python
dy_da = dy_dd * dd_da
print(dy_da)
```

```python
dy_db = (dy_dd * dd_db) + (dy_de * de_db)
print(dy_db)
```

```python
dy_dc = dy_de * de_dc
print(dy_dc)
```

### Parte B, realizamos el cálculo de gradientes con PyTorch

Pista: no necesitas instalar nada, ya viene instalado en Colab!

```python
import torch
```

Ejercicio 11.3 Inicializa las entradas.

Cada nodo de entrada debe ser un tensor, y ya que vamos a hacer la gradiente cada tensor debe tener "requires_grad" = True

Pista: https://pytorch.org/docs/stable/tensors.html

Pista 2: https://pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html


```python
a = torch.tensor(1.0, requires_grad = True)
b = torch.tensor(2.0, requires_grad = True)
c = torch.tensor(3.0, requires_grad = True)
```

```python
d = a + b
e = b - c
```

```python
y = d * e
```

```python
y.backward()
```

```python
a.grad
```

```python
b.grad
```

```python
c.grad
```
