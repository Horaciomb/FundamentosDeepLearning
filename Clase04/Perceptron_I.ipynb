## Perceptron

```python
# Parmetros de entrenamiento
lr = 0.01
epochs = 1000

# Definir datos de entrenamiento para la ecuacin 2w + 3u = 17
X_train = np.array([[2, 3]])
y_train = np.array([17])
```

### **Ejercicio B.1, Inicializa los pesos randmicos**

Pista: No incluyas el bias

Pista 2: Usa random.rand de numpy

```python
# Inicializacin de los pesos
weights = np.random.rand(2)
```

```python
weights
```

```python
X_train.shape
```

```python
# Entrenamiento del perceptrn simple
for epoch in range(epochs):
  for i in range(X_train.shape[0]):
    prediction = np.dot(X_train[i], weights)
    error = y_train[i] - prediction
    weights = weights + lr * error * X_train[i]
  print("EPOCH:", epoch + 1)
  print("Pesos en este epoch:", weights)
```

Comprueba los resultados:

```python
# Mostrar los valores de w y u encontrados
print(weights)
w, u = weights
```

Realiza el calculo inicial con los pesos encontrados

```python
#2w + 3u = 17
print(2*w + 3*u)
```
